```python
# you are free to use any package you deem fit

import sklearn
from sklearn.model_selection import train_test_split
```

## Dataset

```python
# DO NOT MODIFY
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, Y = make_moons(random_state=42, n_samples=(50, 450), noise=0.25)

# take a look at the data
# plot one moon as red and the other as blue
for coordinate, color in zip(X, Y):
    point_color = "r" if color == 0 else "b"
    plt.scatter(coordinate[0], coordinate[1], c=point_color, label=color)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)

print(x_test.shape)
```

## Training

```python
# this is a classification problem so import a classifier model
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

model = RandomForestClassifier()
model2 = SVC()
model3 = KNeighborsClassifier()

model.fit(x_train, y_train)
model2.fit(x_train, y_train)
model3.fit(x_train, y_train)
```

## Testing / Optimization

```python
print(model.score(x_test, y_test))
print(model2.score(x_test, y_test))
print(model3.score(x_test, y_test))
```

This seems to be a relatively easy problem. I can use three different models out of the box with no parameter tuning and consistently get 90 plus percent accuracy with all three.
